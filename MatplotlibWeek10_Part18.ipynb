### Simple 2x2 Layout dengan subplots
- Membuat subplots dengan ukuran 2 panjang dan 2 lebar.
- Parameter Constrained_layout berfungsi untuk menentukann jarak tampilan axes yang satu dengan yang lainnya.
- Untuk mengakses axes kita bisa menggunakan format seperti ini axes [index_row] [index_col]

```python
siswa = ['Cecep', 'Joko', 'Wati', 'Ani']
uas = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True)

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
axes[1][1].barh(siswa, uas)

plt.show()
```

### Simple 2x2 layout dengan GridSpec
- Selain menggunakan subplots, kita juga bisa menggunakan Gridspec.
- Objek gridspec dapat dibentuk dengan memanggil fungsi gridspec.GridSpec().
- Untuk menghubungkan gridspec dengan objek figure kita bisa menambahkan paraemeter figure.
- Untuk membuat objek axes kita bisa memannggil fungsi fig.add_subplot().

```python
fig = plt.figure(constrained_layout=True)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# Plotting
ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

- Cara lain untuk membentuk gridspec adalah memanggil fungsi fig.add_gridspec().
- Fungsi tersebut lebih efisien karena kita tidak perlu lagi menghubungkan antara gridpec dan figurenya

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

### Rows Span dan Cols Span dengan GridSpec
- Untuk melakukan span baris/ kolom kita bisa memasukkan ' : ' pada value yang diinginkan.
- ' : ' berarti menggunakan keseluruhan baris/ kolom yang tersedia.
- Hal ini mirip dengan teknik slicing.
- Jadi selama kita mengerti teknik slicing, kita dapat dengan mumdah memahami rows/ cols span dengan gridpec.s

```python
# Membuat layout basic yang sama dengan layout yang sebelumnya
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

# Membuat axes
ax1 = fig.add_subplot(gs[0, :])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

# ax2 tidak bisa digunakan karena kolom 1 sudah diokupasi oleh ax1
# ax2 = fig.add_subplot(gs[0, 1])
# ax2.pie(uas, labels=siswa)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, :-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')

ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')

plt.show()
```
