#Mengenal Object Oriented Style untuk melakukan plotting

Membuat figures beserta axes, dan memanggil methods dari keduanya

```python
x = np.linspace(0, 2, 100) # (p1, p2, p3) merepresentasikan sebagai (starting_point, n_point, jumlahData_point)
x
```

```python
fig, ax = plt.subplots() 
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x_label') 
ax.set_ylabel('y_label') 
ax.set_title('Simple Plot') 
ax.legend() 
```
